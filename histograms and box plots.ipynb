```python
import pandas as pd
norm_reviews = pd.read_csv("fandango_scores.csv")[["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]]
print(norm_reviews.head(5))
```

```python
import matplotlib.pyplot as plt
from numpy import arange
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
```

```python

fig, ax = plt.subplots()
ax.boxplot(norm_reviews["RT_user_norm"].values)
ax.yaxis.grid(True)
ax.set_ylim(0,5,0.5)
plt.xlabel('Rotten Tomatoes')
plt.show()
```

```python
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.hist(norm_reviews[ 'Fandango_Ratingvalue'], bins=20, range=(0, 5))
ax1.set_xlabel('Fandango')
ax1.set_ylabel('Rotten Tomatoes')

ax2.hist(norm_reviews['RT_user_norm'], bins=20, range=(0, 5))
ax2.set_ylabel('Fandango')
ax2.set_xlabel('Rotten Tomatoes')
plt.show()
```

```python
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews["Fandango_Ratingvalue"], bins=20, range=(0, 5))
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Fandango Ratings")
ax1.set_xlim(0, 5)
ax1.set_ylim(0, 40)

ax2.hist(norm_reviews["Metacritic_user_nom"], bins=20, range=(0, 5))
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_xlim(0, 5)
ax2.set_ylim(0, 30)

ax3.hist(norm_reviews["IMDB_norm"], bins=20, range=(0, 5))
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB Ratings")
ax3.set_xlim(0, 5)
ax3.set_ylim(0, 40)

ax4.hist(norm_reviews['RT_user_norm'], bins=20, range=(0, 5))
ax4.set_xlabel("Fandango")
ax4.set_ylabel("Rotten Tomatoes")
ax4.set_xlim(0, 5)
ax4.set_ylim(0, 20)
plt.show()
```
