```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('unrate.csv')
```

```python
df.DATE = pd.to_datetime(df.DATE)
```

```python

fig = plt.figure(figsize=(14,7))
fig_1948 = fig.add_subplot(2,1,1)
fig_1949 = fig.add_subplot(2,1,2)
fig_1948.plot(df[0:12]['DATE'],df[0:12]['VALUE'])
fig_1949.plot(df[12:24]['DATE'], df[12:24]['VALUE'])
fig_1948.set_title('Monthly Unemployment Rate, 1948')
fig_1949.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

```python
fig_1948_1952 = plt.figure(figsize=(12,14))

for i in range(1953-1948):
    ax= fig_1948_1952.add_subplot(1953-1948,1,i+1)
    ax.plot(df[(i*12):(i+1)*12]['DATE'],df[(i*12):(i+1)*12]['VALUE'])
    ax.set_title('Monthly Unemployment Rate, 19'+str(48+i))
plt.show()
```

```python
fig2_1948_1952 = plt.figure(figsize=(10,6))
for i in range(1953-1948):
    plt.plot(df[(i*12):(i+1)*12]['DATE'].dt.month, df[(i*12):(i+1)*12]['VALUE'], label=str(1948+i))

plt.title('Monthly Unemployment Trends, 1948-1952')
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.legend(loc='upper right')    
plt.show()
```
